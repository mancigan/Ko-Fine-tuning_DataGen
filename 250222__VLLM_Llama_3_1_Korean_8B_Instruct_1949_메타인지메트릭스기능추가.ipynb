```python
# ğŸ“Œ Colab ì‹¤í–‰ìš© ì½”ë“œ: ì‚¬ë¶„ë©´ ë¶„ì„ ì‹œìŠ¤í…œ

# 1ï¸âƒ£ í•„ìˆ˜ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ì¹˜ (Colabì—ì„œëŠ” ì²˜ìŒ ì‹¤í–‰ ì‹œ í•„ìš”)
!pip install vllm transformers pandas --quiet

# 2ï¸âƒ£ ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¶ˆëŸ¬ì˜¤ê¸°
from vllm import LLM, SamplingParams
from transformers import AutoTokenizer, AutoModel
import pandas as pd
import torch  # âœ… NameError ë°©ì§€ë¥¼ ìœ„í•´ ì¶”ê°€

# 3ï¸âƒ£ AI ëª¨ë¸ ë¡œë“œ
BASE_MODEL = "sh2orc/Llama-3.1-Korean-8B-Instruct"
llm = LLM(model=BASE_MODEL, dtype="float16")  # ğŸ“Œ í•˜ë‚˜ì˜ ëª¨ë¸ì„ ëª¨ë“  ì—ì´ì „íŠ¸ê°€ ê³µìœ 

tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = 'right'

# 4ï¸âƒ£ AI ê¸°ë°˜ í…ìŠ¤íŠ¸ ìƒì„± í•¨ìˆ˜
def gen(instruction):
    """AI ëª¨ë¸ì„ í™œìš©í•˜ì—¬ ë¶„ì„ì„ ìˆ˜í–‰í•˜ëŠ” í•¨ìˆ˜"""
    messages = [
        {
            "role": "system",
            "content": "ë‹¹ì‹ ì€ AI ë¶„ì„ ì—ì´ì „íŠ¸ì…ë‹ˆë‹¤. ë°ì´í„°ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë¶„ì„í•˜ê³  ëª©í‘œë¥¼ ì„¤ì •í•˜ì„¸ìš”."
        },
        {
            "role": "user",
            "content": instruction
        },
    ]

    prompt_message = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True,
    )

    eos_token_id = [tokenizer.eos_token_id, tokenizer.convert_tokens_to_ids("<|eot_id|>")]

    outputs = llm.generate(prompt_message, SamplingParams(
        stop_token_ids=eos_token_id,
        temperature=0.3,
        top_p=0.9,
        frequency_penalty=1.3,
        max_tokens=4096))

    return outputs[0].outputs[0].text


# ğŸ“Œ í† í° ì„ë² ë”© ëª¨ë¸ (VLLMì€ ì„ë² ë”© ìƒì„± ë¶ˆê°€ëŠ¥í•˜ë¯€ë¡œ ë³„ë„ ëª¨ë¸ ì‚¬ìš©)
EMBEDDING_MODEL = "jhgan/ko-sbert-multitask"
embedding_tokenizer = AutoTokenizer.from_pretrained(EMBEDDING_MODEL)
embedding_model = AutoModel.from_pretrained(EMBEDDING_MODEL)



# ğŸ”¹ íƒìƒ‰ëœ ë°ì´í„° ì €ì¥ìš© DataFrame ìƒì„±
exploration_data = pd.DataFrame(columns=["Domain", "Explored Actions"])



def clean_and_split_actions(actions):
    """ë¶ˆí•„ìš”í•œ ë²ˆí˜¸ ì œê±° & í–‰ë™ ë°ì´í„°ì™€ ì„¤ëª… ë°ì´í„° ë¶„ë¦¬"""

    cleaned_actions = []
    descriptions = {}

    for action in actions:
        # 1ï¸âƒ£ ë¹ˆ ê°’ ì œê±°
        if not action.strip():
            continue

        # 2ï¸âƒ£ ë¶ˆí•„ìš”í•œ ë©”íƒ€ë°ì´í„° ì œê±°
        exclude_keywords = ["ê¸°ì¡´ í–‰ìœ„ì˜ ë³€í˜•", "ê°™ì€ ì˜ë¯¸ì§€ë§Œ ë‹¤ë¥¸ í‘œí˜„", "ìµœì‹  íŠ¸ë Œë“œ ë°˜ì˜", "ìƒˆë¡œìš´ hÃ nhìœ„"]
        if any(keyword in action for keyword in exclude_keywords):
            continue

        # 3ï¸âƒ£ "VS" ê°œë… ì œê±° (ë¹„êµ ê°œë…)
        if "vs" in action.lower():
            continue

        # 4ï¸âƒ£ ì•ì— ë¶™ì€ ë²ˆí˜¸ ì œê±° (ì˜ˆ: "1.", "2.", "3. **ì œëª©**: ì„¤ëª…")
        action = re.sub(r"^\d+\.\s*", "", action).strip()

        # 5ï¸âƒ£ "**ì œëª©**: ì„¤ëª…" í˜•ì‹ì˜ ë°ì´í„° ë¶„ë¦¬
        if ": " in action:
            title, desc = action.split(": ", 1)
            cleaned_actions.append(title.strip())  # í–‰ë™ ë°ì´í„°ë§Œ ë‚¨ê¹€
            descriptions[title.strip()] = desc.strip()  # ì„¤ëª…ì€ ë³„ë„ ì €ì¥
        else:
            cleaned_actions.append(action.strip())  # ì„¤ëª…ì´ ì—†ìœ¼ë©´ ê·¸ëŒ€ë¡œ ì €ì¥

    return list(set(cleaned_actions)), descriptions  # ì¤‘ë³µ ì œê±° í›„ ë°˜í™˜


def get_token_embeddings(text):
    """í† í° ì„ë² ë”©ì„ ìƒì„± (VLLMì€ ì„ë² ë”© ë¶ˆê°€ â†’ Sentence Transformer í™œìš©)"""
    tokens = embedding_tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        token_embeddings = embedding_model(**tokens).last_hidden_state
    return token_embeddings.mean(dim=1).squeeze()  # ë¬¸ì¥ ì„ë² ë”© ë°˜í™˜

def cosine_similarity(vec1, vec2):
    """ì½”ì‚¬ì¸ ìœ ì‚¬ë„ ê³„ì‚°"""
    return torch.nn.functional.cosine_similarity(vec1.unsqueeze(0), vec2.unsqueeze(0)).item()

def detect_action_changes(existing_actions, generated_actions, threshold=0.8):
    """ê¸°ì¡´ í–‰ë™ê³¼ ì˜ë¯¸ê°€ ìœ ì‚¬í•˜ì§€ë§Œ í‘œí˜„ì´ ë‹¤ë¥¸ í–‰ë™ì„ íƒìƒ‰ (VLLM ê¸°ë°˜)"""
    modified_actions = []
    existing_embeddings = {action: get_token_embeddings(action) for action in existing_actions}

    for action in generated_actions:
        action_embedding = get_token_embeddings(action)
        for existing_action, existing_embedding in existing_embeddings.items():
            similarity = cosine_similarity(action_embedding, existing_embedding)
            if similarity > threshold and action != existing_action:
                modified_actions.append(action)
                break  # ì´ë¯¸ ìœ ì‚¬í•œ í–‰ë™ì´ ë°œê²¬ë˜ì—ˆìœ¼ë¯€ë¡œ ì¶”ê°€ íƒìƒ‰ ì¤‘ë‹¨

    return list(set(modified_actions))  # ì¤‘ë³µ ì œê±° í›„ ë°˜í™˜


# ğŸ”¹ ê¸°ì¡´ íƒìƒ‰ëœ ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜¤ëŠ” í•¨ìˆ˜
def load_explored_actions(domain):
    """ì €ì¥ëœ íƒìƒ‰ëœ í–‰ë™ ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜´"""
    try:
        df = pd.read_json("exploration_data.json")
        actions = df[df["Domain"] == domain]["Explored Actions"].explode().tolist()
        return actions if actions else []
    except (FileNotFoundError, ValueError):
        return []

# ğŸ”¹ íƒìƒ‰ëœ ë°ì´í„°ë¥¼ ì €ì¥í•˜ëŠ” í•¨ìˆ˜
def save_explored_actions(domain, actions, descriptions):
    """íƒìƒ‰ëœ í–‰ë™ ë°ì´í„°ë¥¼ ì¶•ì í•˜ì—¬ ì €ì¥"""
    global exploration_data

    # ğŸ”¹ ê¸°ì¡´ ë°ì´í„°ì™€ í•©ì³ ì¤‘ë³µ ì œê±°
    unique_actions = list(set(actions))

    # ğŸ”¹ DataFrame ì—…ë°ì´íŠ¸
    new_data = pd.DataFrame({"Domain": [domain], "Explored Actions": [unique_actions], "Descriptions": [descriptions]})
    exploration_data = pd.concat([exploration_data, new_data], ignore_index=True)

    # ğŸ”¹ ë°ì´í„° ì €ì¥ (íŒŒì¼ë¡œ ì¶•ì )
    exploration_data.to_json("exploration_data.json", orient="records", force_ascii=False)


# 5ï¸âƒ£ íƒìƒ‰ ì—ì´ì „íŠ¸ T (ì „ì²´ í–‰ë™ íƒìƒ‰)

def explore_actions(domain):
    """íƒìƒ‰ ì—ì´ì „íŠ¸: íŠ¹ì • ì˜ì—­ì—ì„œ ìˆ˜í–‰í•  ìˆ˜ ìˆëŠ” í–‰ë™ì„ íƒìƒ‰"""

    # ğŸ“Œ 1ï¸âƒ£ ê¸°ë³¸ ì œê³µ ë¦¬ìŠ¤íŠ¸ (ì‚¬ì „ ì •ì˜ëœ í–‰ë™)
    predefined_actions = {
        "ë‹¤ì´ì–´íŠ¸": [
            "ì¹¼ë¡œë¦¬ ê³„ì‚°", "ì‹ë‹¨ ê³„íš ìˆ˜ë¦½", "ì‹ì‚¬ëŸ‰ ì¡°ì ˆ", "ê°€ê³µì‹í’ˆ ì¤„ì´ê¸°",
            "ë‹¨ë°±ì§ˆ ì„­ì·¨ ëŠ˜ë¦¬ê¸°", "ìˆ˜ë¶„ ì„­ì·¨ëŸ‰ ì¦ê°€", "ì €ë… ëŠ¦ê²Œ ë¨¹ì§€ ì•Šê¸°",
            "ìŒì‹ ê¸°ë¡í•˜ê¸°", "ë°°ê³ í”Œ ë•Œ ê³¼ì‹ ë°©ì§€", "ì™¸ì‹ ë©”ë‰´ ì¡°ì ˆ"
        ]
    }

    # ğŸ“Œ 2ï¸âƒ£ ì‚¬ì „ ì •ì˜ëœ ë¦¬ìŠ¤íŠ¸ ì œê³µ
    actions = predefined_actions.get(domain, [])

    # ğŸ“Œ 3ï¸âƒ£ ê¸°ì¡´ íƒìƒ‰ëœ í–‰ë™ ë°ì´í„° ê°€ì ¸ì˜¤ê¸°
    existing_actions = load_explored_actions("ë‹¤ì´ì–´íŠ¸") or []

    # ğŸ“Œ 4ï¸âƒ£ AI íƒìƒ‰ ìˆ˜í–‰ (ìƒˆë¡œìš´ í–‰ë™ & ê¸°ì¡´ í–‰ë™ ë³€í™” íƒìƒ‰)
    instruction = f"""
    '{domain}' ì˜ì—­ì—ì„œ ìˆ˜í–‰í•  ìˆ˜ ìˆëŠ” ëª¨ë“  í–‰ë™ì„ ë‚˜ì—´í•˜ì„¸ìš”.
    - ê¸°ì¡´ ë°ì´í„°ì— ì—†ëŠ” ìƒˆë¡œìš´ í–‰ë™ì„ ìš°ì„ ì ìœ¼ë¡œ íƒìƒ‰í•˜ì„¸ìš”.
    - ê¸°ì¡´ í–‰ë™ì˜ ë³€í˜• ë˜ëŠ” ìµœì‹  íŠ¸ë Œë“œë¥¼ ë°˜ì˜í•œ ë³€í™”ë¥¼ íƒìƒ‰í•˜ì„¸ìš”.
    - ê°™ì€ ì˜ë¯¸ì´ì§€ë§Œ ë‹¤ë¥´ê²Œ í‘œí˜„ë  ìˆ˜ ìˆëŠ” í–‰ë™ì´ ìˆëŠ”ì§€ í™•ì¸í•˜ì„¸ìš”.
    """
    generated_actions = gen(instruction)

    # ğŸ”¹ 4ï¸âƒ£ AI ê²°ê³¼ê°€ ë¬¸ìì—´ì´ë©´ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜
    if isinstance(generated_actions, str):
        generated_actions = generated_actions.split("\n")  # ì¤„ë°”ê¿ˆ ê¸°ì¤€ìœ¼ë¡œ ë¦¬ìŠ¤íŠ¸ ë³€í™˜

    # ğŸ”¹ 6ï¸âƒ£ ê¸°ì¡´ ë°ì´í„°ì™€ ë¹„êµí•˜ì—¬ **ìƒˆë¡œìš´ í–‰ë™ë§Œ í•„í„°ë§**
    new_actions = list(set(generated_actions) - set(existing_actions))

    # ğŸ”¹ 7ï¸âƒ£ ê¸°ì¡´ í–‰ë™ì˜ ë³€í™”ë‚˜ ìƒˆë¡œìš´ íŠ¸ë Œë“œ ë°˜ì˜ ì—¬ë¶€ í™•ì¸ (ë³€í™” íƒìƒ‰)
    modified_actions = detect_action_changes(existing_actions, generated_actions)

    # ğŸ“Œ 8ï¸âƒ£ ë¶ˆí•„ìš”í•œ ë°ì´í„° ì œê±° ë° ì„¤ëª… ë¶„ë¦¬
    cleaned_actions, action_descriptions = clean_and_split_actions(new_actions + modified_actions)

    # ğŸ“Œ 9ï¸âƒ£ ê¸°ì¡´ ë¦¬ìŠ¤íŠ¸ + AI íƒìƒ‰ ê²°ê³¼ ë°˜í™˜
    all_actions = actions + cleaned_actions

    # ğŸ”¹ 10ï¸âƒ£ íƒìƒ‰ëœ ë°ì´í„° ì €ì¥ (ì¤‘ë³µ ì œê±° í›„)
    save_explored_actions(domain, all_actions, action_descriptions)




# 6ï¸âƒ£ ê¸°ì¤€ Aë¥¼ ë¶„ë¥˜í•˜ëŠ” ì—ì´ì „íŠ¸ A
def classify_A(actions):
    """ ì•ˆë‹¤/ëª¨ë¥¸ë‹¤"""
    instruction = f"ì•ˆë‹¤/ëª¨ë¥¸ë‹¤ë¡œ êµ¬ë¶„í•˜ì„¸ìš”: {actions}"
    return gen(instruction)


# 7ï¸âƒ£ ê¸°ì¤€ Bë¥¼ ë¶„ë¥˜í•˜ëŠ” ì—ì´ì „íŠ¸ B
def classify_B(actions):
    """í• ìˆ˜ìˆë‹¤/í• ìˆ˜ì—†ë‹¤."""
    instruction = f"í• ìˆ˜ìˆë‹¤/í• ìˆ˜ì—†ë‹¤ë¡œ êµ¬ë¶„í•˜ì„¸ìš”: {actions}"
    return gen(instruction)


# 8ï¸âƒ£ ëª©í‘œ ì„¤ì • ì—ì´ì „íŠ¸ G
def define_goal(actions_A, actions_B):
    """ìµœê³ ìˆ˜ì¤€ì„ ì„¤ì •"""
    instruction = f"ì •í•´ì§„ ì˜ì—­ì—ì„œ ìµœê³ ìˆ˜ì¤€ì„ íƒìƒ‰/ì„¤ì •í•˜ì„¸ìš”."
    return gen(instruction)


# 9ï¸âƒ£ ì‚¬ìš©ìê°€ ì›í•˜ëŠ” ë¶„ì„ ì˜ì—­ ì„ íƒ ê¸°ëŠ¥
def select_domain():
    """ì‚¬ìš©ìê°€ ì›í•˜ëŠ” ë¶„ì„ ì˜ì—­ì„ ì„ íƒí•  ìˆ˜ ìˆë„ë¡ ì œê³µ"""
    available_domains = ["ìê¸°ê³„ë°œ", "ë‹¤ì´ì–´íŠ¸", "ìš´ë™", "í•™ìŠµ", "ê¸°íƒ€"]
    print("\nğŸ“Œ ë¶„ì„í•  ì˜ì—­ì„ ì„ íƒí•˜ì„¸ìš”:")
    for idx, domain in enumerate(available_domains, 1):
        print(f"{idx}. {domain}")

    choice = int(input("\nğŸ”¹ ì„ íƒ (ë²ˆí˜¸ ì…ë ¥): "))

    if choice == len(available_domains):  # "ê¸°íƒ€"ë¥¼ ì„ íƒí•˜ë©´ ì‚¬ìš©ì ì…ë ¥ ë°›ìŒ
        domain = input("\nğŸ”¹ ì§ì ‘ ì…ë ¥í•˜ì„¸ìš” (ì˜ˆ: ì—…ë¬´ íš¨ìœ¨, ì·¨ë¯¸ ê°œë°œ ë“±): ")
    else:
        domain = available_domains[choice - 1]

    print(f"\nâœ… ì„ íƒëœ ì˜ì—­: {domain}")
    return domain

# ğŸ“Œ Colabì—ì„œ `matplotlib` ì‹¤í–‰ ì˜¤ë¥˜ ë°©ì§€
%matplotlib inline


# 9ï¸âƒ£ ë¶„ì„ ê²°ê³¼ë¥¼ ë§¤íŠ¸ë¦­ìŠ¤(ì‚¬ë¶„ë©´)ì— ë°°ì¹˜í•˜ëŠ” í•¨ìˆ˜
def plot_matrix(actions_A, actions_B):
    """ê¸°ì¤€ A, Bë¥¼ ê¸°ë°˜ìœ¼ë¡œ ì‚¬ë¶„ë©´(ë§¤íŠ¸ë¦­ìŠ¤) ê·¸ë˜í”„ë¥¼ ìƒì„±"""

    # Xì¶•: ê³„íš ì—¬ë¶€ (B)
    # Yì¶•: ì‹¤ì²œ ì—¬ë¶€ (A)
    fig, ax = plt.subplots(figsize=(6, 6))

    # ì‚¬ë¶„ë©´ ì˜ì—­ ì„¤ì •
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)

    # ì¶• ë¼ë²¨ ì„¤ì •
    ax.set_xlabel("ê³„íš ì—¬ë¶€ (B)")
    ax.set_ylabel("ì‹¤ì²œ ì—¬ë¶€ (A)")

    # ì¤‘ì•™ì„  ì¶”ê°€
    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)

    # ì‚¬ë¶„ë©´ ì œëª© ì„¤ì •
    ax.text(0.5, 0.8, "1ì‚¬ë¶„ë©´\n(ì‹¤ì²œ ìˆìŒ, ê³„íš ìˆìŒ)", fontsize=12, ha='center')
    ax.text(-0.5, 0.8, "2ì‚¬ë¶„ë©´\n(ì‹¤ì²œ ìˆìŒ, ê³„íš ì—†ìŒ)", fontsize=12, ha='center')
    ax.text(-0.5, -0.8, "3ì‚¬ë¶„ë©´\n(ì‹¤ì²œ ì—†ìŒ, ê³„íš ì—†ìŒ)", fontsize=12, ha='center')
    ax.text(0.5, -0.8, "4ì‚¬ë¶„ë©´\n(ì‹¤ì²œ ì—†ìŒ, ê³„íš ìˆìŒ)", fontsize=12, ha='center')

    # ì‚¬ë¶„ë©´ì— í–‰ë™ ë°°ì¹˜
    for action in actions_A.split(", "):
        if action in actions_B:
            ax.text(0.5, 0.5, action, fontsize=10, ha='center', color='blue')  # 1ì‚¬ë¶„ë©´
        elif action not in actions_B:
            ax.text(-0.5, 0.5, action, fontsize=10, ha='center', color='red')  # 2ì‚¬ë¶„ë©´

    for action in actions_B.split(", "):
        if action not in actions_A:
            ax.text(0.5, -0.5, action, fontsize=10, ha='center', color='green')  # 4ì‚¬ë¶„ë©´
        elif action not in actions_B:
            ax.text(-0.5, -0.5, action, fontsize=10, ha='center', color='purple')  # 3ì‚¬ë¶„ë©´

    plt.title("ì‚¬ë¶„ë©´ ë¶„ì„ ë§¤íŠ¸ë¦­ìŠ¤")
    plt.show()



# ğŸ”Ÿ Colabì—ì„œ ì‹¤í–‰í•  ë©”ì¸ í•¨ìˆ˜
def main():
    """Colabì—ì„œ ì‹¤í–‰í•  ì „ì²´ ë¶„ì„ ê³¼ì •"""
    print("âœ… ë‹¤ë‹¨ê³„ ì—ì´ì „íŠ¸ ê¸°ë°˜ ì‚¬ë¶„ë©´ ë¶„ì„ ì‹œìŠ¤í…œ ì‹¤í–‰ ì¤‘...")

    # ğŸ”¹ ì‚¬ìš©ì ì…ë ¥ìœ¼ë¡œ ë¶„ì„í•  ì˜ì—­ ì„ íƒ
    domain = select_domain()

    print(f"\nğŸ” '{domain}' ì˜ì—­ì˜ í–‰ë™ì„ íƒìƒ‰ ì¤‘...")
    all_actions = explore_actions(domain)
    print(f"ğŸ“Œ íƒìƒ‰ëœ í–‰ë™ ëª©ë¡:\n{all_actions}")

    print("\nğŸ›  ê¸°ì¤€ A (ì‹¤ì²œ ì—¬ë¶€)ë¡œ í–‰ë™ì„ ë¶„ë¥˜ ì¤‘...")
    classified_A = classify_A(all_actions)
    print(f"ğŸ“Œ ê¸°ì¤€ Aì— ë”°ë¥¸ ë¶„ë¥˜ ê²°ê³¼:\n{classified_A}")

    print("\nğŸ›  ê¸°ì¤€ B (ê³„íš ì—¬ë¶€)ë¡œ í–‰ë™ì„ ë¶„ë¥˜ ì¤‘...")
    classified_B = classify_B(all_actions)
    print(f"ğŸ“Œ ê¸°ì¤€ Bì— ë”°ë¥¸ ë¶„ë¥˜ ê²°ê³¼:\n{classified_B}")

    print("\nğŸ¯ ëª©í‘œ ì„¤ì • ì¤‘...")
    goal = define_goal(classified_A, classified_B)
    print(f"ğŸš€ ì„¤ì •ëœ ëª©í‘œ:\n{goal}")

    print("\nâœ… ë¶„ì„ ì™„ë£Œ! ëª¨ë“  ê²°ê³¼ê°€ ì¶œë ¥ë˜ì—ˆìŠµë‹ˆë‹¤.")

# ğŸ”Ÿ ë©”ì¸ ì‹¤í–‰
main()
```
